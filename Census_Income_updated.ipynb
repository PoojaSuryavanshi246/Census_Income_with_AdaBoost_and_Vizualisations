```python
for x in df.columns:
    if df[x].dtype != 'object':
        plt.boxplot(df[x])
        plt.xlabel(x)
        plt.show()
```

```python
out_list = ['hours-per-week', 'education-num', 'fnlwgt']

for x in out_list:
    Q1 = df[x].quantile(0.25)
    Q3 = df[x].quantile(0.75)
    IQR = Q3 - Q1
    lb = Q1 - 1.5 * IQR
    ub = Q3 + 1.5 * IQR
    df = df[(df[x] >= lb) & (df[x] <= ub)]
```

```python
df = df[[col for col in df.columns if col != 'annual_income'] + ['annual_income']]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['annual_income'].value_counts()
```

```python
df['annual_income'] = le.fit_transform(df['annual_income'])
print(le.classes_, le.transform(le.classes_))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X = df.iloc[:, :-1]
y = df['annual_income']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 20)
```

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(X_train, y_train)
```

```python
log_ypred = log_reg.predict(X_test)
```

```python
accuracy_score(log_ypred, y_test)
```

```python
lrl1 = LogisticRegression(penalty='l1', solver='liblinear')
```

```python
lrl1.fit(X_train, y_train)
```

```python
lrl1_ypred = lrl1.predict(X_test)
print(accuracy_score(lrl1_ypred, y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train, y_train)
```

```python
dt_ypred = dt.predict(X_test)
```

```python
accuracy_score(dt_ypred, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators = 100, random_state = 20)
```

```python
rf.fit(X_train, y_train)
```

```python
rf_ypred = rf.predict(X_test)
```

```python
accuracy_score(rf_ypred, y_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ada = AdaBoostClassifier(n_estimators = 220, random_state = 20)
```

```python
ada.fit(X_train, y_train)
```

```python
ada_ypred = ada.predict(X_test)
```

```python
accuracy_score(ada_ypred, y_test)
```

```python
# Shift each model vertically + add jitter
offsets = [0.00, 0.02, 0.04, 0.06, 0.08, 0.10]  # Actual + 5 models
names = ['Actual', 'Logistic', 'Logistic with L1', 'DT', 'RF', 'AdaBoost']
preds = [y_test.values, log_ypred, lrl1_ypred, dt_ypred, rf_ypred, ada_ypred]
colors = ['black','yellow', 'orange', 'green', 'red', 'purple']

plt.figure(figsize=(12, 6))

for offset, pred, name, col in zip(offsets, preds, names, colors):
    x = np.arange(len(pred))
    y = pred + offset + np.random.normal(0, 0.02, len(pred))  # jitter + offset
    plt.scatter(x, y, label=name, color=col, alpha=0.6, s=8)

plt.title('Actual vs All Models (Separated for Clarity)')
plt.xlabel('Sample Index')
plt.ylabel('Label (+ small offset)')
plt.ylim(-0.05, 1.3)
plt.legend()
plt.grid(True, alpha=0.2)
plt.show()
```

```python
print("Accuracy:")
print("Logistic:", accuracy_score(y_test, log_ypred))
print("Logistic L1:", accuracy_score(y_test, lrl1_ypred))
print("DT:", accuracy_score(y_test, dt_ypred))
print("RF:", accuracy_score(y_test, rf_ypred))
print("AdaBoost:", accuracy_score(y_test, ada_ypred))  # â† highest = best
```

```python
from sklearn.metrics import f1_score

scores = {
    'Logistic': accuracy_score(y_test, log_ypred),
    'Logistic L1': accuracy_score(y_test, lrl1_ypred),
    'DT': accuracy_score(y_test, dt_ypred),
    'RF': accuracy_score(y_test, rf_ypred),
    'AdaBoost': accuracy_score(y_test, ada_ypred)
}

plt.bar(scores.keys(), scores.values(), color=['orange', 'green', 'red', 'purple'])
plt.title('Model Accuracy Comparison')
plt.ylabel('Accuracy')
for i, v in enumerate(scores.values()):
    plt.text(i, v + 0.005, f"{v:.3f}", ha='center')
plt.ylim(0, 1)
plt.show()
```

```python

from sklearn.metrics import confusion_matrix

models = ['Logistic', 'Logistic with L1', 'Decision Tree', 'Random Forest', 'AdaBoost']
preds = [log_ypred, lrl1_ypred, dt_ypred, rf_ypred, ada_ypred]

fig, axes = plt.subplots(1, 5, figsize=(16, 4))

for i, (pred, name) in enumerate(zip(preds, models)):
    cm = confusion_matrix(y_test, pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[i], cbar=False)
    axes[i].set_title(f'{name}')
    axes[i].set_xlabel('Predicted')
    axes[i].set_ylabel('Actual')

plt.tight_layout()
plt.show()
```
